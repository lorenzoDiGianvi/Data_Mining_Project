```python
# to show an image automatically within the browser without using show()
%matplotlib inline 

import math
import numpy as np # for working with arrays
import pandas as pd # for working with tables
import scipy.stats as stats
import matplotlib.pyplot as plt # for visualization (creating plots)

import statistics as sts # statistic module for variance
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
```

```python
df = pd.read_csv('8_drop_categorical.csv')
```

```python
from pandas.plotting import scatter_matrix
```

```python
dfnew = df.drop(columns=['Attrition','EnvironmentSatisfaction', 'Gender','Education' ,'JobInvolvement','JobSatisfaction','NumCompaniesWorked','OverTime','RelationshipSatisfaction','StockOptionLevel','TrainingTimesLastYear','MaritalStatus' 'WorkLifeBalance' ])
```

```python
dfnew.describe()
```

```python
scatter_matrix(dfnew, figsize=(10, 10))
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(dfnew.values)
```

```python
X[:5]
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
kmeans = KMeans(n_clusters=3, n_init=10, max_iter=100)
kmeans.fit(X)
```

```python
np.unique(kmeans.labels_, return_counts=True)
```

```python
centers = scaler.inverse_transform(kmeans.cluster_centers_)
plt.scatter(dfnew['Stagnation'], dfnew['YearsAtCompany'], c=kmeans.labels_, 
            s=20)
plt.scatter(centers[:, 0], centers[:, 3], s=200, marker='*', c='k')
plt.tick_params(axis='both', which='major', labelsize=22)
plt.show()
```
